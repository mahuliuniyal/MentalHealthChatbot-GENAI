```python
from langchain.embeddings import HuggingFaceBgeEmbeddings
from langchain.document_loaders import PyPDFLoader, PyPDFDirectoryLoader
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_groq import ChatGroq
import os
import gradio as gr

def initialise_llm():
    llm = ChatGroq(
        temperature=0,
        groq_api_key="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",
        model_name="llama-3.3-70b-versatile"
    )
    return llm

def create_vector_db():
    loader = PyPDFLoader("/content/mental_health_Document.pdf") # Corrected file path
    documents = loader.load()
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
    texts = text_splitter.split_documents(documents)
    embeddings = HuggingFaceBgeEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
    vector_db = Chroma.from_documents(texts, embeddings, persist_directory='./chroma_db')
    vector_db.persist()
    print("ChromaDB created and data saved")
    return vector_db

def setup_qa_chain(vector_db, llm):
    retriever = vector_db.as_retriever()
    prompt_template = """You are a passionate mental health chatbot. Respond thoughtfully to the following question:
    {context}
    User: {question}
    Chatbot:"""
    PROMPT = PromptTemplate(template=prompt_template, input_variables=['context', 'question'])
    qa_chain = RetrievalQA.from_chain_type(
        llm=llm,
        chain_type="stuff",
        retriever=retriever,
        chain_type_kwargs={"prompt": PROMPT}
    )
    return qa_chain

def main():
    print("Initialising Chatbot...........")
    llm = initialise_llm()
    db_path = "./chroma_db"

    if not os.path.exists(db_path):
        vector_db = create_vector_db()
    else:
        embeddings = HuggingFaceBgeEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
        vector_db = Chroma(persist_directory=db_path, embedding_function=embeddings)

    qa_chain = setup_qa_chain(vector_db, llm)

    def chatbot_response(message, history):
        if not message.strip():
            return "Please provide a valid input"

        try:
            response = qa_chain({"query": message})
            return response["result"]
        except Exception as e:
            return f"Error: {str(e)}"

    with gr.Blocks(theme='gradio/soft') as app:
        gr.Markdown("# Mental Health Chatbot")
        gr.Markdown("Ask me anything about mental health and I'll help you based on the knowledge I have.")

        chatbot = gr.ChatInterface(
            fn=chatbot_response,
            title="Mental Health Support",
            description="A supportive chatbot for mental health discussions"
        )

    app.launch(debug=True, share=True)


if __name__=="__main__":
  main()
```
